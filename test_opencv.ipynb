```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("img/group_poop.png")
print(img)
print(img.shape) # This image is a huge array!
print('//////')
print(img[30][30]) # This is a pixel
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img) #Matplotlib likes things in RBG like a sensible person
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
print(img.shape)
plt.imshow(img, cmap='gray')
```

```python
img = img[80:430, 75:470] #Array slicing
plt.imshow(img, cmap='gray')
```

```python
sobelx = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)
plt.imshow(sobelx, cmap='gray')
```

```python
sobely = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=9)
plt.imshow(sobely, cmap='gray')
```
